## Words

```python
words = ['play', 'filling', 'bar', 'theatre', 'easygoing', 'date', 'lead', 'that', 'story',  'island', 'trying to survive to git']
```

**Print every word in upper case**

```python
for i in words:
    print (i)
```

**Create a new list containing only words with 5 or more letters**

```python
for i in words:
    if len(i)>=5:
        print(i)
```

```python
for i in words:
    if i.startswith('t'):
        print (i)
          
```

## People

```python
people = [
    {
        "name": "Juan",
        "age": 34,
        "n_kids": 2
    },
    {
        "name": "Pepe",
        "age": 27,
        "n_kids": 0
    },
    {
        "name": "Sonia",
        "age": 41,
        "n_kids": 1
    },
    {
        "name": "Lucía",
        "age": 22,
        "n_kids": 2
    },
    {
        "name": "Leo",
        "age": 55,
        "n_kids": 5
    }
]
```

**How many people are there?**

```python
len(people)
```

**How many people have kids**?

```python
count = 0
for i in people:
    if i["n_kids"] == 0:
        count += 1

print(count)  
```

**How many kids do they have in total?**

```python
total_kids = 0
for i in people:
    if i["n_kids"] > 0:
        total_kids += i["n_kids"]

print(total_kids) 
```

**In a year's time, names ending with "a" will have an extra kid. Create a list of dictionaries with people's info in a year's time**

```python
people_next_year = []

for i in people:
    new_people_dict = i.copy()  
    if new_people_dict["name"].endswith("a"):
        new_people_dict["n_kids"] += 1
    people_next_year.append(new_people_dict)

print(people_next_year)
```
